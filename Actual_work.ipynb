```python
import pandas as pd
import numpy as np
```

```python
file_schools = 'schools.csv'
file_students = 'students.csv'
```

```python
school_df = pd.read_csv(file_schools)
students_df = pd.read_csv(file_students)
```

```python
school_df.head()
```

```python
students_df.head()
```

```python
school_data_complete = pd.merge(school_df, students_df, how='left', on = 'school_name')
school_data_complete.head()
```

```python
#schoolcount
school_count = school_data_complete["school_name"].unique()
real_school_count = len(school_count)
real_school_count
```

```python
#student count
district_student_count = school_df["size"].sum()
district_student_count
```

```python
#budget
budget = school_df["budget"].sum()
budget
```

```python
#average math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
#average reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
#overall passing students
overall_passing_students = students_df[(students_df["math_score"] >= 70) & (students_df["reading_score"] >= 70)]["student_name"].count()
overall_passing_students
#overall passing average
overall_percent_passing = overall_passing_students/students_df["student_name"].count()
```

```python
#students passing math
students_passing_math = students_df[(students_df["math_score"] >= 70)]["student_name"].count()
students_passing_math
percent_passing_math = students_passing_math/students_df["student_name"].count()
percent_passing_math
```

```python
#students passing reading
students_passing_reading = students_df[(students_df["reading_score"] >= 70)]["student_name"].count()
students_passing_reading
percent_passing_reading = students_passing_reading/students_df["student_name"].count()
percent_passing_reading
```

```python
district_summary = pd.DataFrame({
    "Total Schools": [real_school_count],
    "Total Students": [district_student_count],
    "Total Budget": [budget],
    "Average Reading Score": [average_reading_score],
    "Average Math Score": [average_math_score],
    "% Passing Reading": [percent_passing_reading],
    "% Passing Math": [percent_passing_math],
    "Overall Passing Rate": [overall_percent_passing]
})
district_summary
```

```python
#FINAL DISTRICT SUMMARY

district_summary.style.format({
    "Total Budget": "${:,.2f}",
    "Average Reading Score": "{:.1f}",
    "Average Math Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "Overall Passing Rate": "{:.1%}"
})
```

```python
data_by_school = school_data_complete.set_index('school_name').groupby('school_name')
data_by_school.head()
```

```python
school_types = school_df.set_index('school_name')['type']
school_types
```

```python
#student per school
student_per_school = data_by_school["Student ID"].count()
student_per_school
```

```python
#budget_per_school
budget_by_school = school_df.set_index("school_name")['budget']
budget_by_school
```

```python
#PER STUDENT BUDGET
per_student_budget = school_df.set_index("school_name")['budget']/school_df.set_index("school_name")['size']
per_student_budget
```

```python
#average math score
average_math_by_school = data_by_school['math_score'].mean()
average_math_by_school
```

```python
#average reading score
average_reading_by_school = data_by_school['reading_score'].mean()
average_reading_by_school
```

```python
# % passing math by school
grouped_passing_math = students_df[students_df["math_score"]>=70].groupby('school_name')['Student ID'].count()
percent_passing_math_by_school = grouped_passing_math/student_per_school
percent_passing_math_by_school
```

```python
# % passing reading by school
grouped_passing_reading = students_df[students_df['reading_score']>=70].groupby('school_name')['Student ID'].count()
percent_passing_reading_by_school = grouped_passing_reading/student_per_school
percent_passing_reading_by_school
```

```python
#OVERALL PASSING RATE
overall_passing_rate_by_school = ((percent_passing_math_by_school) + (percent_passing_reading_by_school))/2
overall_passing_rate_by_school
```

```python
everything_by_school = pd.DataFrame({
    "School Types": school_types,
    "Total Students": student_per_school,
    "Total School Budget": budget_by_school,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_by_school,
    "Average Reading Score": average_reading_by_school,
    "% Passing Math": percent_passing_math_by_school,
    "% Passing Reading": percent_passing_reading_by_school,
    "Overall Passing Rate": overall_passing_rate_by_school
})
```

```python
everything_by_school.style.format({
    "Total Students": "{:,}",
    "Total School Budget": "{:,}",
    "Per Student Budget": "${:.0f}",
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "Overall Passing Rate": "{:.1%}"
})
```

```python
top_5 = everything_by_school.sort_values("Overall Passing Rate", ascending = False)
real_top_5 = top_5.head().style.format({
    "Total Students": "{:,}",
    "Total School Budget": "{:,}",
    "Per Student Budget": "${:.0f}",
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "Overall Passing Rate": "{:.1%}"
})
real_top_5
```

```python
bottom_5 = everything_by_school.nsmallest(5, 'Overall Passing Rate')
bottom_5.style.format({
    "Total Students": "{:,}",
    "Total School Budget": "{:,}",
    "Per Student Budget": "${:.0f}",
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "Overall Passing Rate": "{:.1%}"
})
```

```python
#ninth grade 
ninth_grade_math = students_df.loc[students_df['grade']=='9th'].groupby('school_name')["math_score"].mean()
#tenth grade math
tenth_grade_math = students_df.loc[students_df['grade']=='10th'].groupby('school_name')["math_score"].mean()
#11th grade
eleventh_grade_math = students_df.loc[students_df['grade']=='11th'].groupby('school_name')["math_score"].mean()
#12th
twelth_grade_math = students_df.loc[students_df['grade']=='12th'].groupby('school_name')["math_score"].mean()
```

```python
#mathscores data frame
math_scores_by_grade = pd.DataFrame({
    "9th": ninth_grade_math,
    "10th": tenth_grade_math,
    "11th": eleventh_grade_math,
    "12th": twelth_grade_math
})

math_scores_by_grade.style.format({
    "9th": "{:.1f}",
    "10th":"{:.1f}",
    "11th":"{:.1f}",
    "12th":"{:.1f}"
})
```

```python
#reading scores
#ninth grade 
ninth_grade_reading = students_df.loc[students_df['grade']=='9th'].groupby('school_name')["reading_score"].mean()
#tenth grade math
tenth_grade_reading = students_df.loc[students_df['grade']=='10th'].groupby('school_name')["reading_score"].mean()
#11th grade
eleventh_grade_reading = students_df.loc[students_df['grade']=='11th'].groupby('school_name')["reading_score"].mean()
#12th
twelth_grade_reading = students_df.loc[students_df['grade']=='12th'].groupby('school_name')["reading_score"].mean()
```

```python
#readingscores data frame
reading_scores_by_grade = pd.DataFrame({
    "9th": ninth_grade_reading,
    "10th": tenth_grade_reading,
    "11th": eleventh_grade_reading,
    "12th": twelth_grade_reading
})

reading_scores_by_grade.style.format({
    "9th": "{:.1f}",
    "10th":"{:.1f}",
    "11th":"{:.1f}",
    "12th":"{:.1f}"
})
```

```python
#spending bin shit 
bins = [0, 550, 600, 650, 1100]
spending_group_names = ["$0-$550", "$550-$600", "$600-$650", "$650-$1100"]
school_data_complete["spending_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels = spending_group_names)
school_data_complete
```

```python
group_by_spending_bins = school_data_complete.groupby("spending_bins")
```

```python
average_math_by_spending = group_by_spending_bins["math_score"].mean()
average_math_by_spending
```

```python
average_reading_by_spending = group_by_spending_bins["reading_score"].mean()
average_reading_by_spending
```

```python
number_passing_math_spending = school_data_complete[school_data_complete["math_score"]>=70].groupby("spending_bins")["Student ID"].count()
percent_passing_math_spending = number_passing_math_spending/ group_by_spending_bins["Student ID"].count()
```

```python
number_passing_reading_spending = school_data_complete[school_data_complete["reading_score"]>=70].groupby("spending_bins")["Student ID"].count()
percent_passing_reading_spending = number_passing_reading_spending/ group_by_spending_bins["Student ID"].count()
```

```python
overall_passing_spending = ((percent_passing_reading_spending) + (percent_passing_math_spending))/2
```

```python
scores_by_spend = pd.DataFrame({
    "Average Math Score": average_math_by_spending,
    "Average Reading Score": average_reading_by_spending,
    "% Passing Math": percent_passing_math_spending,
    "% Passing Reading": percent_passing_reading_spending,
    "Overall Passing Rate": overall_passing_spending
})
scores_by_spend.index.name = "Per Student Budget"
scores_by_spend.style.format({
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "Overall Passing Rate": "{:.1%}"
})
```

```python
#sschool size
bins = [0, 1500,3000,40000]
school_size_classification = ["Small(<1500)", "Medium(1500-3000)", "Large(>3000)"]
school_data_complete["student_population_size"] = pd.cut(school_data_complete["size"], bins, labels = school_size_classification)
school_data_complete
```

```python
group_by_school_size_classification = school_data_complete.groupby("student_population_size")
```

```python
average_math_score_population = group_by_school_size_classification["math_score"].mean()
```

```python
average_reading_score_population = group_by_school_size_classification["reading_score"].mean()
```

```python
number_passing_math_size_classification = school_data_complete[school_data_complete["math_score"]>=70].groupby("student_population_size")["Student ID"].count()
percent_passing_math_size = number_passing_math_size_classification/ group_by_school_size_classification["Student ID"].count()
```

```python
number_passing_reading_size_classification = school_data_complete[school_data_complete["reading_score"]>=70].groupby("student_population_size")["Student ID"].count()
percent_passing_reading_size = number_passing_reading_size_classification/ group_by_school_size_classification["Student ID"].count()
```

```python
overall_passing_size_classification = ((percent_passing_math_size) + (percent_passing_reading_size))/2
```

```python
scores_by_size = pd.DataFrame({
    "Average Math Score": average_math_score_population,
    "Average Reading Score": average_reading_score_population,
    "% Passing Math": percent_passing_math_size,
    "% Passing Reading": percent_passing_reading_size,
    "Overall Passing Rate": overall_passing_size_classification
})
scores_by_size.index.name = "Student Population Size"
scores_by_size.style.format({
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.1%}",
    "% Passing Reading": "{:.1%}",
    "Overall Passing Rate": "{:.1%}"
})
```
